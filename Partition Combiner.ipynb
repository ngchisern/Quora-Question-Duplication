```python
import numpy as np
import pandas as pd
```

```python
dfs_left = []
dfs_right = []

dfs_left.append(pd.read_csv("0_60000_left.csv"))
dfs_right.append(pd.read_csv("0_60000_right.csv"))
```

```python
start = 60001
end = 360001
window = 60000

for i in range(start, end, window):
    print(i)
    left = "{}_{}_left.csv".format(i, i + window - 1)
    right = "{}_{}_right.csv".format(i, i + window - 1)
    
    dfs_left.append(pd.read_csv(left))
    dfs_right.append(pd.read_csv(right))
```

```python
df_left = pd.concat(dfs_left, ignore_index=True)
```

```python
df_left
```

```python
df_right = pd.concat(dfs_right, ignore_index=True)
```

```python
df_right
```

```python
df_left.to_csv("left_all.csv", index=False)
```

```python
df_right.to_csv("right_all.csv", index=False)
```
