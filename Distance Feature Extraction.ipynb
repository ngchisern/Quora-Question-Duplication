```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('cleaned_dataset.csv', index_col=0)
```

```python
df_bert_1 = pd.read_csv('left_final.csv')
df_bert_2 = pd.read_csv('right_final.csv')
```

```python
from scipy.spatial import distance

bert1 = df_bert_1.to_numpy()
bert2 = df_bert_2.to_numpy()
```

```python
def apply(func, left, right):
    result = []
    
    for i in range(len(left)):
        if (i % 50000 == 0):
            print(i, "th iteration")
        
        result.append(func(left[i], right[i]))
    
    return result
```

```python
chebyshev = apply(distance.chebyshev, bert1, bert2)
```

```python
df['chebyshev'] = chebyshev
```

```python
braycurtis = apply(distance.braycurtis, bert1, bert2)
```

```python
df['braycurtis'] = braycurtis
```

```python
cosine = apply(distance.cosine, bert1, bert2)
```

```python
df['cosine'] = cosine
```

```python
def stats(obj):
    return obj.statistic

correlation = list(map(stats, apply(pearsonr, bert1, bert2)))
```

```python
df['correlation'] = correlation
```

```python
canberra = apply(distance.canberra, bert1, bert2)
```

```python
df['canberra'] = canberra
```

```python
cityblock = apply(distance.cityblock, bert1, bert2)
```

```python
df['cityblock'] = cityblock
```

```python
def getEuclidean(left, right):
    result = []
    
    for i in range(len(left)):
        if (i % 50000 == 0):
            print(i, "th iteration")
        result.append(np.linalg.norm(left[i] - right[i]))
    return result

euclidean = getEuclidean(bert1, bert2)
```

```python
df['euclidean'] = euclidean
```

```python
def getMinkowski(left, right, p):
    result = []
    
    for i in range(len(left)):
        if (i % 50000 == 0):
            print(i, "th iteration")
        result.append(distance.minkowski(left[i], right[i], p))
    return result

minkowski = getMinkowski(bert1, bert2, 3)
```

```python
df['minkowski'] = minkowski
```

```python
from nltk.corpus import stopwords
from nltk import download
import gensim.downloader as api

download('stopwords')  # Download stopwords list.
stop_words = stopwords.words('english')
model = api.load('word2vec-google-news-300')
```

```python
def preprocess(sentence):
    return [w for w in sentence.lower().split() if w not in stop_words]

def getMoverDistance(sentence1, sentence2):    
    return model.wmdistance(sentence1, sentence2)
```

```python
def generateMoverDistance(data):
    result = []
    
    for i in range(len(data)):
        if (i % 50000 == 0):
            print(i, "th iteration")
        
        result.append(getMoverDistance(data[i][3], data[i][4]))
    
    return result
```

```python
data_list = df.to_numpy()
```

```python
df['word_mover_distance'] = generateMoverDistance(data_list)
```

```python
from scipy.stats import skew
```

```python
def getSkewness(data):
    result = []
    
    for i in range(len(data)):
        if (i % 50000 == 0):
            print(i, "th iteration")
        
        result.append(skew(data[i], bias=True))
    
    return result
```

```python
df['skew_q1'] = getSkewness(bert1)
```

```python
df['skew_q2'] = getSkewness(bert2)
```

```python
def getKurtosis(data):
    result = []
    
    for i in range(len(data)):
        if (i % 50000 == 0):
            print(i, "th iteration")
        
        result.append(kurtosis(data[i], bias=True))
    
    return result
```
