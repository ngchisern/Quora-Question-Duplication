```python
import numpy as np
import pandas as pd
```

```python
series = list(range(0,768))
left_names = ['q1_' + str(i) for i in series]
right_names = ['q2_' + str(i) for i in series]
```

```python
start = 1
end = 20001
window = 5000
```

```python
dfs_left = []
dfs_right = []

for i in range(start, end, window):
    left = "{}_{}_left.csv".format(i - 1, i + window - 1)
    right = "{}_{}_right.csv".format(i - 1, i + window - 1)
    
    dfs_left.append(pd.read_csv(left))
    dfs_right.append(pd.read_csv(right))
```

```python
dfs_left.append(pd.read_csv("20000_45000_left.csv"))
dfs_right.append(pd.read_csv("20000_45000_right.csv"))
```

```python
dfs_left.append(pd.read_csv("45000_60000_left.csv"))
dfs_right.append(pd.read_csv("45000_60000_right.csv"))
```

```python
df_left = pd.concat(dfs_left, ignore_index=True)
```

```python
df_right = pd.concat(dfs_right, ignore_index=True)
```

```python
left_file_name = "{}_{}_left.csv".format(start, end - 1)
df_left.to_csv(left_file_name, index=False)
```

```python
right_file_name = "{}_{}_right.csv".format(start, end - 1)
df_right.to_csv(right_file_name, index=False)
```
