```python
def process_and_clean_data(dataframe):
  # Convert to lower case
  dataframe.question1=dataframe.question1.str.lower()
  dataframe.question2=dataframe.question2.str.lower()
  # Change label at rows with noise to 1 instead of removing,
  # since 1 is minority class
  dataframe.iloc[NOISE_ROW_IDX, dataframe.columns.get_loc('is_duplicate')] = 1
  return dataframe
```

```python
df=pd.read_csv(GLUE_URL)
```

```python
processed_and_cleaned_df = process_and_clean_data(df)
```
